#### 2.1 Aplica una transformaci√≥n puntual. Modifica el valor de los pixeles considerando un valor de umbral (threshold). la funci√≥n de treshold debe ser una rampa con una pendiente dada (tu defines el valor de la pendiente) y cuyo valor m√≠nimo ser√° 0 y el m√°ximo ser√° 255 (valores t√≠picos de los pixeles).

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_threshold_ramp(image, threshold, slope):
    # Crear una funci√≥n de rampa con la pendiente y el umbral dados
    def ramp(x):
        return np.clip(slope * (x - threshold) + 127.5, 0, 255)
    
    # Aplicar la funci√≥n de rampa a cada p√≠xel de la imagen
    transformed_image = ramp(image)
    
    return transformed_image.astype(np.uint8)

# Cargar una imagen en escala de grises
image = cv2.imread("imagen_7.png", cv2.IMREAD_GRAYSCALE)

# Definir el umbral y la pendiente
threshold = 55
slope = 1.5

# Aplicar la transformaci√≥n
transformed_image = apply_threshold_ramp(image, threshold, slope)

# Mostrar la imagen original y la transformada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagen Transformada')
plt.imshow(transformed_image, cmap='gray')
plt.axis('off')

plt.show()
```

#### 2.2 Aplica una transformaci√≥n en la vecindad. 
- Debes hacer una vecindad con dimensiones de 5x5 pixeles. 
- Aplica la transformaci√≥n de tal forma que los pixeles cambien para tener el valor correspondiente al m√°ximo de todos los vecinos.
- Debes hacer la misma operaci√≥n a lo largo y ancho de toda la imagen. ESto implica, de forma iterativa, recorrer la imagen.



```python
"""
Para aplicar una transformaci√≥n en la vecindad con dimensiones de 5x5 p√≠xeles y ajustar cada p√≠xel
para que tome el valor m√°ximo de todos sus vecinos, estamos hablando de una operaci√≥n de filtrado
que se conoce como dilataci√≥n en procesamiento de im√°genes. Esta operaci√≥n pertenece a la categor√≠a 
de morfolog√≠a matem√°tica.
"""

import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_max_filter(image, kernel_size):
    # Aplicar la operaci√≥n de dilataci√≥n
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    transformed_image = cv2.dilate(image, kernel, iterations=1)
    
    return transformed_image

# Cargar una imagen en escala de grises
image = cv2.imread("imagen_7.png", cv2.IMREAD_GRAYSCALE)

# Definir el tama√±o de la vecindad (kernel_size)
kernel_size = 5

# Aplicar la transformaci√≥n
transformed_image = apply_max_filter(image, kernel_size)

# Mostrar la imagen original y la transformada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagen Transformada')
plt.imshow(transformed_image, cmap='gray')
plt.axis('off')

plt.show()
```

#### 2.3 Aplica transformaci√≥n de intensidad

- Debes hacer una vecindad con dimensiones de 10x10 pixeles. 
- Aplica la transformaci√≥n que aparece en la ecuaci√≥n de tal forma que los pixeles cambien dependiendo de la posici√≥n (r) y el coeficiente c (de tu elecci√≥n).
- Debes hacer la misma operaci√≥n a lo largo y ancho de toda la imagen. ESto implica, de forma iterativa, recorrer la imagen.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def log_transform(image, kernel_size, c):
    # Obtener las dimensiones de la imagen
    rows, cols = image.shape
    
    # Crear una imagen de salida
    transformed_image = np.zeros((rows, cols), dtype=np.float32)
    
    # Definir el radio m√°ximo
    max_r = kernel_size // 2
    
    # Iterar sobre cada p√≠xel de la imagen
    for i in range(max_r, rows - max_r):
        for j in range(max_r, cols - max_r):
            # Extraer la vecindad de 10x10
            neighborhood = image[i-max_r:i+max_r+1, j-max_r:j+max_r+1]
            
            # Aplicar la transformaci√≥n logar√≠tmica a cada p√≠xel de la vecindad
            for m in range(neighborhood.shape[0]):
                for n in range(neighborhood.shape[1]):
                    r = neighborhood[m, n]
                    transformed_image[i, j] = c * np.log(1 + r)
    
    # Normalizar la imagen transformada a valores entre 0 y 255
    transformed_image = np.clip(transformed_image, 0, 255)
    transformed_image = transformed_image.astype(np.uint8)
    
    return transformed_image

# Cargar una imagen en escala de grises
image = cv2.imread("imagen_7.png", cv2.IMREAD_GRAYSCALE)

# Definir el tama√±o de la vecindad y el coeficiente c
kernel_size = 10
c = 3  # Puede ajustarse seg√∫n sea necesario

# Aplicar la transformaci√≥n
transformed_image = log_transform(image, kernel_size, c)

# Mostrar la imagen original y la transformada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagen Transformada')
plt.imshow(transformed_image, cmap='gray')
plt.axis('off')

plt.show()
```

#### 2.4 Operaci√≥n con Kernel
- Realiza tres operaciones de convoluci√≥n en cascada. 
- Utiliza torch nn.Conv2d
- Debes calcular el stride, padding y dem√°s par√°metros para que obtengas 16 canales de salida.
- Muestra una imagen de cada canal resultante de la convoluci√≥n.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n para mostrar una imagen
def show_image(image, title=''):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Cargar una imagen en escala de grises y normalizarla
image = cv2.imread("imagen_7.png", cv2.IMREAD_GRAYSCALE)
image = image.astype(np.float32) / 255.0

# Convertir la imagen a un tensor de PyTorch
image_tensor = torch.from_numpy(image).unsqueeze(0).unsqueeze(0)  # Shape: (1, 1, H, W)

# Definir una secuencia de capas de convoluci√≥n
class ConvCascade(nn.Module):
    def __init__(self):
        super(ConvCascade, self).__init__()
        self.conv1 = nn.Conv2d(1, 8, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(8, 16, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(16, 16, kernel_size=3, stride=1, padding=1)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        return x

# Inicializar el modelo y aplicar la convoluci√≥n en cascada
model = ConvCascade()
output = model(image_tensor)

# Mostrar cada uno de los 16 canales resultantes
output_np = output.detach().numpy().squeeze()  # Shape: (16, H, W)

for i in range(output_np.shape[0]):
    show_image(output_np[i], title=f'Canal {i+1}')
```

## 4. Segmentaci√≥n de imagen
#### 4.1 Realiza un algoritmo para dividir una imagen con 3 canales RGB de 512x512, en parches (secciones) de 16x16. Presenta cada parche generado de la imagen.
El algoritmo debe tolerar el cambio de imagen por una de menor tamano (128, 64). Claramente, en cada caso el n√∫mero de parches ser√° menor. 
    

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def segmentar_imagen(imagen, size=16):
    parches = []
    altura, ancho, _ = imagen.shape
    
    for y in range(0, altura, size):
        for x in range(0, ancho, size):
            parche = imagen[y:y+size, x:x+size]
            parches.append(parche)
    
    return parches

# Cargar la imagen
imagen = cv2.imread("Imagen_gatos.jpeg")

# Verificar que la imagen est√° cargada correctamente
if imagen is None:
    raise ValueError("No se pudo cargar la imagen. Verifica el path.")

# Segmentar la imagen en parches de 16x16
parches = segmentar_imagen(imagen, size=16)

# Mostrar cada parche generado
num_parches = len(parches)
num_columnas = 8  # Para mostrar 8 parches por fila
num_filas = num_parches // num_columnas + (num_parches % num_columnas > 0)

plt.figure(figsize=(10, 10))
for i, parche in enumerate(parches):
    plt.subplot(num_filas, num_columnas, i + 1)
    plt.imshow(cv2.cvtColor(parche, cv2.COLOR_BGR2RGB))
    plt.axis('off')

plt.show()
```
